## 1. What is QuaRCS?

The QuaRCS (Quantitative Reasoning for College Science) Assessment is a scientific and quantitative literacy assessment for general education science students. The assessment serves as a tool to guage math anxiety in college-level students and determine the contributions instructors have to the math performance of their students. The assessment is administered at the beginning and end of each semester to mark the progress of students who participate in the assessment. The assesment features a series of demographic and attitudinal questions at the end as well as confidence questions for each quantitative question. Originally, the assessment featured 25 questions, but more recently has offered 15 questions. The assesment has been administered at research universities, public and private, liberal arts colleges and community colleges all across the nation.
